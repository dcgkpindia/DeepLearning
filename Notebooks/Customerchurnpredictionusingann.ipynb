### Some explanation of Standard Scaler class because of Question -
<b> Why we used fit_transform to train set and transform to test set? </b>

The fit_transform() method and the transform() method in the StandardScaler class are both used to standardize data. However, they have different purposes.

The fit_transform() method first fits the scaler to the data, and then transforms the data. This means that the scaler will learn the mean and standard deviation of the data in the fit_transform() method, and then use those values to transform the data in the transform() method.

The transform() method, on the other hand, simply transforms the data without fitting the scaler to the data first. This means that the scaler must be fitted to the data before it can be used to transform the data using the transform() method.

In general, the fit_transform() method should be used when you want to standardize the data and then use the scaler to transform the data again later. The transform() method should be used when you only need to transform the data once.

The example

This is because the StandardScaler class standardizes data by subtracting the mean and then dividing by the standard deviation.

## Model

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(16, activation = 'relu', input_dim = 11))
model.add(Dense(12, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
opt = keras.optimizers.Adam(learning_rate = 0.01)
```

```python
model.compile(loss="binary_crossentropy", optimizer = opt, metrics=['accuracy', 'Precision', 'Recall', 'AUC'])
```

```python
history = model.fit(X_train_scaled, y_train, batch_size=50, epochs=100,verbose=1,validation_split=0.2 )
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis = 1)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
